```python
import pandas as pd
import numpy as np
from scipy.fft import fft
from statsmodels.tsa.seasonal import seasonal_decompose

def dft_past(signal):
    N = len(signal)
    X = np.zeros(N, dtype=complex)
    
    for k in range(N):
        sum_value = 0
        for n in range(k):
            sum_value += signal[n] * np.exp(-1j * 2 * np.pi * k * n / N)
        X[k] = sum_value
    
    return X

def featureextraction(df):
    df = df.sort_values(by=['Customer_ID', 'Month'])
    
    feature_list = []
    
    for customer_id, group in df.groupby('Customer_ID'):
        group = group.copy()
        
        for i in range(1, 4):
            group[f'lag_{i}'] = group['Amount_Unpaid'].shift(i)
        
        group['rolling_mean'] = group['Amount_Unpaid'].rolling(window=3).mean().shift(1)
        group['rolling_std'] = group['Amount_Unpaid'].rolling(window=3).std().shift(1)
        
        try:
            decomposition = seasonal_decompose(group['Amount_Unpaid'].fillna(0), model='additive', period=3)
            group['trend'] = decomposition.trend.shift(1)
            group['seasonal'] = decomposition.seasonal.shift(1)
            group['residual'] = decomposition.resid.shift(1)
        except ValueError as e:
            print(f"Erreur dans la décomposition saisonnière pour le client {customer_id}: {e}")
        
        fft_values = dft_past(group['Amount_Unpaid'].fillna(0).to_numpy())
        fft_real = fft_values.real
        fft_imag = fft_values.imag
        
        if len(fft_real) < len(group):
            fft_real = np.concatenate([fft_real, np.nan * np.zeros(len(group) - len(fft_real))])
            fft_imag = np.concatenate([fft_imag, np.nan * np.zeros(len(group) - len(fft_imag))])
        
        group['fft_real'] = fft_real
        group['fft_imag'] = fft_imag
        
        group['Amount_Unpaid'] = group['Amount_Unpaid']
        
        feature_list.append(group)
    
    df_features = pd.concat(feature_list)
    
    datafinale = df_features[['Customer_ID', 'Month', 'Amount_Unpaid', 'lag_1', 'lag_2', 'lag_3', 
                                        'rolling_mean', 'rolling_std', 'trend', 'seasonal', 'residual', 'fft_real', 'fft_imag']]
    
    return datafinale

```
